```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cat = pd.read_csv("/content/category_margin.csv")
subcat = pd.read_csv("/content/subcat_margin.csv")
loss = pd.read_csv("/content/loss_products.csv")
region = pd.read_csv("/content/regional_pref.csv")
monthly = pd.read_csv("/content/monthly_sales.csv")
inventory = pd.read_csv("/content/inventory_profit.csv")
```

```python
print("Category Margin:\n", cat)
print("\nSub-Category Margin:\n", subcat.head())
print("\nTop Loss-Making Products:\n", loss.head())
print("\nRegional Performance:\n", region)
print("\nMonthly Sales & Profit:\n", monthly.head())
print("\nInventory vs Profit:\n", inventory.head())
```

```python
corr = inventory[['avg_inventory_days','profit_margin']].corr().iloc[0,1]
print("\nCorrelation (Inventory Days vs Profit Margin):", corr)
```

```python
plt.scatter(inventory['avg_inventory_days'], inventory['profit_margin'])
plt.xlabel("Average Inventory Days")
plt.ylabel("Profit Margin")
plt.title("Inventory Days vs Profitability")
plt.show()
```

```python
plt.savefig("Inventory Days vs Profitability.png")
```

```python
monthly['year_month'] = pd.to_datetime(monthly['year_month'])
plt.plot(monthly['year_month'], monthly['total_sales'], label="Sales")
plt.plot(monthly['year_month'], monthly['total_profit'], label="Profit")
plt.legend()
plt.title("Monthly Sales and Profit Trend")
plt.show()
```

```python
plt.savefig("Monthly Sales and Profit Trend.png")
```

```python
plt.bar(cat['category'], cat['total_profit'])
plt.title("Profit by Category")
plt.show()
```

```python
plt.savefig("Profit by Category.png")
```

```python
subcat_sorted = subcat.sort_values('profit_margin')
plt.barh(subcat_sorted['sub_category'], subcat_sorted['profit_margin'])
plt.title("Profit Margin by Sub-Category")
plt.show()
```

```python
plt.savefig("Profit Margin by Sub-Category.png")
```
